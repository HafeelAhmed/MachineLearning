## Importing the necessary packages

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import requests as rq
```

## Get the Respose status of the URL

```python
url ="https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
request = rq.get(url)
request
```

## Get the HTML Source code text

```python
html = bs(request.text)
```

## Extracting Crucial Data
#### This step is vital. Here, we use the beautifulSoup tool to specifically locate and fetch the necessary information.
#### The gathered data is then stored in a variable for later use.

```python
titles = html.select("._4rR01T")
price = html.select("._1_WHN1")
description = html.select(".fMghEO")
offerPercentage= html.select("._3tbKJL span")
```

## Create dataFrame using the information

```python
laptopDetails = pd.DataFrame(titles,columns=["Laptop Name"])
laptopDetails["Price"] = price
laptopDetails["Offer %"] = offerPercentage
laptopDetails["Description"] = description
```

```python
laptopDetails.to_csv("LaptopDetails.csv")
```
