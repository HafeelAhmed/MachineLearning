## how to create a DataFrame usint pandas

```python
season = {"winter":{1:"nov",2:"dec",3:"jan","last":"feb"},
          "Dry":{1:"mar",2:"apr",3:"may","last":"jun"},
          "Spring":{1:"jul",2:"aug",3:"sep","last":"oct"}}
```
